```python
vitals = pd.read_parquet(BASE_DIR / 'vitals.parquet')
vitals.head()
```

```python
# Pivot vitals by type and summarize daily stats
# Keeps resident_id + measured_at in the pivoted base table

vitals_work = vitals.copy()
vitals_work['measured_at'] = pd.to_datetime(vitals_work['measured_at'], errors='coerce')
vitals_work = vitals_work.dropna(subset=['resident_id', 'measured_at', 'vital_type'])

# Ensure numeric values before pivoting/aggregating
vitals_work['value'] = pd.to_numeric(vitals_work['value'], errors='coerce')
vitals_work['dystolic_value'] = pd.to_numeric(vitals_work['dystolic_value'], errors='coerce')

# One row per resident + timestamp, columns per vital type
vitals_pivot = (
    vitals_work
    .pivot_table(
        index=['resident_id', 'measured_at'],
        columns='vital_type',
        values='value',
        aggfunc='mean'
    )
    .reset_index()
)

# Add average dystolic value at the same resident/timestamp grain
dystolic_at_time = (
    vitals_work
    .groupby(['resident_id', 'measured_at'], as_index=False)['dystolic_value']
    .mean()
)

vitals_pivot = vitals_pivot.merge(dystolic_at_time, on=['resident_id', 'measured_at'], how='left')

# Weekly summary per resident: mean and std for each pivoted vital column + mean dystolic_value
vitals_pivot['week_start'] = vitals_pivot['measured_at'].dt.to_period('W').dt.start_time
metric_cols = [c for c in vitals_pivot.columns if c not in ['resident_id', 'measured_at', 'week_start']]

agg_map = {col: ['mean', 'std'] for col in metric_cols if col != 'dystolic_value'}
agg_map['dystolic_value'] = ['mean']

weekly_vitals_stats = (
    vitals_pivot
    .groupby(['resident_id', 'week_start'])
    .agg(agg_map)
)

# Flatten multi-index columns
weekly_vitals_stats.columns = [
    f"{col}_{stat}" if stat else str(col)
    for col, stat in weekly_vitals_stats.columns
]
weekly_vitals_stats = weekly_vitals_stats.reset_index().sort_values(['resident_id', 'week_start'])

print('Pivoted shape:', vitals_pivot.shape)
print('Weekly summary shape:', weekly_vitals_stats.shape)
display(vitals_pivot.head())
display(weekly_vitals_stats.head())

```

```python
adl_work = adl_responses.copy()
adl_work['assessment_date'] = pd.to_datetime(adl_work['assessment_date'], errors='coerce').dt.floor('D')
adl_work = adl_work.dropna(subset=['resident_id', 'assessment_date', 'activity', 'category'])

adl_work['response'] = pd.to_numeric(adl_work['response'], errors='coerce')
adl_work['adl_change'] = pd.to_numeric(adl_work['adl_change'], errors='coerce')

status_clean = adl_work['response_status'].astype(str).str.strip().str.lower()
adl_work['is_complete'] = status_clean.eq('complete').astype('int8')
adl_work['is_non_complete'] = (~status_clean.eq('complete')).astype('int8')

adl_base = (
    adl_work
    .groupby(['resident_id', 'assessment_date', 'activity', 'category'], as_index=False)
    .agg({
        'response': 'mean',
        'adl_change': 'mean',
        'is_complete': 'mean',
        'is_non_complete': 'max'
    })
)


response_wide = adl_base.pivot_table(
    index=['resident_id', 'assessment_date'],
    columns=['activity', 'category'],
    values='response',
    aggfunc='mean'
)

adl_change_wide = adl_base.pivot_table(
    index=['resident_id', 'assessment_date'],
    columns=['activity', 'category'],
    values='adl_change',
    aggfunc='mean'
)

complete_rate_wide = adl_base.pivot_table(
    index=['resident_id', 'assessment_date'],
    columns=['activity', 'category'],
    values='is_complete',
    aggfunc='mean'
)

non_complete_flag_wide = adl_base.pivot_table(
    index=['resident_id', 'assessment_date'],
    columns=['activity', 'category'],
    values='is_non_complete',
    aggfunc='max'
)

# Flatten columns with explicit metric prefix
response_wide.columns = [f"response_mean__{a}__{c}" for a, c in response_wide.columns]
adl_change_wide.columns = [f"adl_change_mean__{a}__{c}" for a, c in adl_change_wide.columns]
complete_rate_wide.columns = [f"status_complete_rate__{a}__{c}" for a, c in complete_rate_wide.columns]
non_complete_flag_wide.columns = [f"status_non_complete_flag__{a}__{c}" for a, c in non_complete_flag_wide.columns]

adl_pivot = (
    response_wide
    .join(adl_change_wide, how='outer')
    .join(complete_rate_wide, how='outer')
    .join(non_complete_flag_wide, how='outer')
    .reset_index()
    .sort_values(['resident_id', 'assessment_date'])
)

print('ADL pivoted shape:', adl_pivot.shape)
display(adl_pivot.head())


```

```python
# Weekly aggregation of ADL pivot features (resident_id + week window)

adl_weekly = adl_pivot.copy()
adl_weekly['assessment_date'] = pd.to_datetime(adl_weekly['assessment_date'], errors='coerce')
adl_weekly = adl_weekly.dropna(subset=['resident_id', 'assessment_date'])

# Weekly windows (Mon-Sun by pandas default weekly period)
adl_weekly['week_start'] = adl_weekly['assessment_date'].dt.to_period('W').dt.start_time
adl_weekly['week_end'] = adl_weekly['assessment_date'].dt.to_period('W').dt.end_time.dt.floor('D')

feature_cols = [c for c in adl_weekly.columns if c not in ['resident_id', 'assessment_date', 'week_start', 'week_end']]

adl_weekly_stats = (
    adl_weekly
    .groupby(['resident_id', 'week_start', 'week_end'], as_index=False)[feature_cols]
    .mean()
    .sort_values(['resident_id', 'week_start'])
)

print('Weekly ADL shape:', adl_weekly_stats.shape)
display(adl_weekly_stats.head())

```
